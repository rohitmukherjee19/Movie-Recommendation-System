# Using Weighted average for each movie's average rating

```python
#Calculate all the components based on the above formula
v = movie_cleaned_df['vote_count']
R = movie_cleaned_df['vote_average']
C = movie_cleaned_df['vote_average'].mean()
m = movie_cleaned_df['vote_count'].quantile(0.7)
```

```python
movie_cleaned_df['weighed_avg'] = ((R*v+C*m)/(v+m))
```

```python
movie_cleaned_df.head()
```

```python
movie_sorted_ranking = movie_cleaned_df.sort_values('weighed_avg', ascending = False)
movie_sorted_ranking[['original_title', 'vote_average', 'vote_count', 'weighed_avg', 'popularity']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average=movie_sorted_ranking.sort_values('weighed_avg',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighed_avg'].head(10), y=weight_average['original_title'].head(10), data=weight_average)
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_popular_movies.png')
```

# Content Based Recommendation System

Now lets make a recommendations based on the movieâ€™s plot summaries given in the overview column. So if our user gives us a movie title, our goal is to recommend movies that share similar plot summaries.

```python
movie_cleaned_df.head(1)['overview']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Filling NaNs with empty string
movie_cleaned_df['overview'] = movie_cleaned_df['overview'].fillna('')
```

```python
# Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(movie_cleaned_df['overview'])
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movie_cleaned_df.index, index=movie_cleaned_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movie_cleaned_df['original_title'].iloc[movie_indices]
```

```python
# Testing our content-based recommendation system with the seminal film Spy Kids
give_rec('Shanghai Calling')
```
